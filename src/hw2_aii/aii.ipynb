```python
import csv
import numpy as np
import pandas as pd
import pyfpgrowth
import time
from datetime import timedelta
```

# 1. Read Data

```python
raw_data = pd.read_csv('../../data/trade_new.csv', index_col=0) # trade_new.csv  and  trade.csv
raw_data = raw_data.reset_index(drop=True)

# raw_data = pd.read_csv('../../data/trade.csv')
# print("Read data...")
```

```python
# only the bndno contains the NaN value
raw_data[np.isnan(raw_data.bndno)]
```

```python
# use a dict to store the transaction of each vip
# also record the sldatime
l_data = len(raw_data)
vip_transaction = {}
vip_set = set()
for i in range(l_data):
    v = raw_data.loc[i, 'vipno']
    t = raw_data.loc[i, 'sldatime'] # sldat -> trade.csv   sldatime -> trade_new.csv
    if v in vip_set:
        vip_transaction[v].append((t, i))
    else:
        vip_transaction[v] = [(t, i)]
        vip_set.add(v)
```

```python
# sort and cut the data
vip_list = list(vip_set)
for v in vip_list:
    # sort
    t = vip_transaction[v]
    t = sorted(t, key=lambda x:x[0])
    
    # cut and pick up the index only
    l = round(len(t) * 0.6)
    t = t[:l]
    vip_transaction[v] = [tt[1] for tt in t] # tt[0] --> time , tt[1] --> index of raw_data
```

```python
def split_transaction(indice, item_label, transactions):
    split_data = raw_data.loc[indice, item_label] # get the uid and pluno/dptno/bndno columns
    if item_label == 'bndno':
        split_data = split_data.dropna(how='any')
        split_data = split_data.astype('int')
    t = list(set(list(split_data))) # remove the duplicate value 
#     t = list(split_data) 
    transactions.append(t)
```

```python
item_labels = ['bndno', 'dptno', 'pluno', ]
time_duration = []
for item_label in item_labels:
    print("item -- {}".format(item_label))
    # fill the transactions with the items
    transactions = []
    for v in vip_list:
        split_transaction(vip_transaction[v], item_label, transactions)
    print(len(transactions)) # print the transaction number
    # call FP algorithm
    supports = [10, 8, 6, 4, 2]
    for s in supports:
        time_start=time.time()
        patterns = pyfpgrowth.find_frequent_patterns(transactions, s)
        with open('fp_{0}_{1}.txt'.format(item_label, s), 'w') as f:
            for p in patterns:
                f.write(str(list(p))+'\n')
        time_end=time.time()
        print('support = {} done. totally cost = '.format(s), timedelta(seconds=time_end-time_start))
        time_duration.append(time_end-time_start)
```

```python
with open('time.csv', 'w') as f:
    t = ','.join([str(x) for x in time_duration[:5]])
    f.write(t + '\n')
    t = ','.join([str(x) for x in time_duration[5:10]])
    f.write(t + '\n')
    t = ','.join([str(x) for x in time_duration[10:]])
    f.write(t + '\n')
```
